# Plotting Functions 

We have included a plotting function to allow users to visually inspect the orbital evolution of a given KBO.  This program requires numpy and matplotlib.

```python
%run plot_backend.ipynb
%matplotlib inline
```

The user must provide a file that contains columns id, time, semi-major axis, inclination, longitude of the ascending node, argument of pericenter, and mean anomaly, in that order.  The rows must contain the information for Neptune (with `id=-1`) and the KBO (with `id=0`). This is the output saved by the functions [`compute_from_jpl()`](KBO Classification.ipynb#Predicting-class-from-Horizons-identifier) and [`compute_from_aei()`](KBO Classification.ipynb#Predicting-class-from-orbital-elements) in the KBO Classification notebook when the keyword `savename` is specified.

### For a six panel plot

To see the semi-major axis, eccentricity, inclination, pericenter distance, longitude of the pericenter, and longitude of the ascending node evolution, specify the filename to the function `plot_obj()`.

```python
plot_obj("K13EF4J.follow")
```

### For a seven panel plot

If you think the body is resonant, you can specify the integers `p`, `q`, `m`, and/or `n` to also plot the resonant angle.  At a minimum, `p` and `q` must be defined.

```python
plot_obj("K04VD0X.follow", p=3, q=2)
```
